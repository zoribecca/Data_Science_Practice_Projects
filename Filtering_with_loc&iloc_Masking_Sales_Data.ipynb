```python
# Filtering/Masking
# Using Booleans "AND" & "OR"
# AND in pandas is written as "&"
# OR in pandas is written as "|"
```

```python
#Filtering
sales[sales["Sales Agent"] == "Blessing"]
```

```python
sales[sales["Sales Agent"] == "Chinedu"]
```

```python
sales[(sales["Sales Agent"] == "Blessing" )| (sales["Sales Agent"] == "Chinedu")]
```

```python
# Masking
# lets pick one of the Sales Agents and see their behaviour
mask = sales["Sales Agent"] == "Blessing"
sales[mask]
```

```python
sales.loc[mask]
```

```python
sales.loc[mask, "Sales Agent":"Price"].head(2)   # Give me mask but from column Sales Agent too Price
```

```python
# Applying multiple filters
# Here we would be using the "&" and "|" operators
# We have picked out the this particular Sales Agent and the units sold that are >= 10
```

```python
mask2 = (sales["Sales Agent"] == "Blessing") & (sales["Units"] >= 10)
sales.loc[mask2]
```

```python
# Let's see the number of units sold that were >=10
this_customer = sales.loc[mask2]
this_customer["Units"].sum()
```

```python
# Or
this_customer = sales.loc[mask2]
this_customer["Units"].value_counts()
```
